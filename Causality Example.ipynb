## Three-step spending-share pipeline

This section turns the exploratory notebook above into a research workflow for small-area household spending estimation.

The empirical object is not individual household expenditure. It is the **expected household spending profile of an area**, conditional on local socioeconomic structure. The unit of analysis is the Local Authority District (LAD).

The recommended modelling sequence is:

1. **Bayesian hierarchical small-area estimation** for prediction and uncertainty.
2. **Double/debiased machine learning (DML)** for selected causal estimands.
3. **GP multi-output model** as a nonlinear multi-output predictive benchmark.

The three phases answer different questions. The Bayesian model estimates partially pooled area spending profiles with uncertainty. The DML layer estimates interpretable conditional causal effects under explicit assumptions. The GP benchmark tests whether a flexible nonlinear multi-output surface improves prediction.


### Research estimands

Let `i` index LADs and `k` index spending categories.

We model two linked outcomes:

```text
spending_intensity_i = total_expenditure_i / median_income_i
composition_share_ik = category_expenditure_ik / sum_k category_expenditure_ik
category_income_share_ik = spending_intensity_i * composition_share_ik
```

This has two advantages:

- `composition_share_i` is a valid composition: every element is positive and the row sums to one.
- `spending_intensity_i` is allowed to be below or above one, reflecting saving, borrowing, definition differences, or measurement differences.

For causal analysis, example estimands are:

```text
ATE of log median income on spending intensity
ATE of deprivation on discretionary-spending composition
Heterogeneous income effects by OAC/persona structure
```

These are observational, area-level causal estimands. They require assumptions; they are not discovered automatically from the CSV.


````python
import os
import numpy as np
import pandas as pd

RANDOM_SEED = 7
np.random.seed(RANDOM_SEED)

DATA_PATH = os.path.join(os.getcwd(), 'NOMIS', 'NOMIS Data 2011 Census.csv')
df_raw = pd.read_csv(DATA_PATH)

# The CSV contains a duplicate raw header. Pandas appends .1 to the second copy.
df_raw = df_raw.loc[:, ~df_raw.columns.str.endswith('.1')].copy()

ID_COLUMNS = ['Local Authority District', 'LAD16CD', 'Region', 'Country']
INCOME_COLUMN = 'Median Income'
TOTAL_SPENDING_COLUMN = 'Total expenditure '

SPENDING_COLUMNS = [
    'Food & non-alcoholic drinks',
    'Alcoholic drink, tobacco & narcotics',
    'Clothing & footwear',
    'Housing (net), fuel & power',
    'Household goods & services',
    'Health',
    'Transport',
    'Communication',
    'Recreation & culture',
    'Education',
    'Restaurants & hotels',
    'Miscellaneous goods & services',
]

for column in SPENDING_COLUMNS + [TOTAL_SPENDING_COLUMN, INCOME_COLUMN]:
    df_raw[column] = pd.to_numeric(df_raw[column], errors='coerce')

analysis_df = df_raw.dropna(subset=SPENDING_COLUMNS + [TOTAL_SPENDING_COLUMN, INCOME_COLUMN]).copy()
analysis_df = analysis_df[analysis_df[INCOME_COLUMN] > 0].copy()
analysis_df = analysis_df.reset_index(drop=True)

print('Raw LAD rows:', len(df_raw))
print('Analysis LAD rows:', len(analysis_df))
analysis_df[ID_COLUMNS + [INCOME_COLUMN, TOTAL_SPENDING_COLUMN]].head()
````

### Outcome construction: compositional spending

The GP and Bayesian models should not directly predict unconstrained category amounts. Spending categories are a composition. We transform the composition with an additive log-ratio (ALR) transform:

```text
z_ik = log(share_ik / share_iK), for k = 1,...,K-1
```

The inverse ALR maps predictions back to positive shares that sum to one. This is the same modelling logic as a logistic-normal composition model.


````python
def safe_log(x, eps=1e-9):
    return np.log(np.clip(np.asarray(x, dtype=float), eps, None))


def alr_transform(shares, eps=1e-9):
    shares = np.clip(np.asarray(shares, dtype=float), eps, 1.0)
    shares = shares / shares.sum(axis=1, keepdims=True)
    reference = shares[:, [-1]]
    return np.log(shares[:, :-1] / reference)


def inverse_alr(z):
    z = np.asarray(z, dtype=float)
    exp_z = np.exp(z)
    denom = 1.0 + exp_z.sum(axis=1, keepdims=True)
    first = exp_z / denom
    last = 1.0 / denom
    return np.hstack([first, last])

category_values = analysis_df[SPENDING_COLUMNS].to_numpy(dtype=float)
category_sum = category_values.sum(axis=1, keepdims=True)
composition = category_values / category_sum

for idx, column in enumerate(SPENDING_COLUMNS):
    analysis_df[f'share__{column}'] = composition[:, idx]

analysis_df['spending_intensity'] = analysis_df[TOTAL_SPENDING_COLUMN] / analysis_df[INCOME_COLUMN]
analysis_df['log_spending_intensity'] = safe_log(analysis_df['spending_intensity'])

Y_alr = alr_transform(composition)
Y_model = np.column_stack([analysis_df['log_spending_intensity'].to_numpy(dtype=float), Y_alr])
Y_OUTPUT_NAMES = ['log_spending_intensity'] + [f'alr__{c}' for c in SPENDING_COLUMNS[:-1]]

print('Composition row-sum range:', composition.sum(axis=1).min(), composition.sum(axis=1).max())
print('Model outcome matrix:', Y_model.shape)
analysis_df[['spending_intensity'] + [f'share__{c}' for c in SPENDING_COLUMNS[:4]]].head()
````

### Feature design and leakage discipline

We use many socioeconomic variables for prediction, but keep feature blocks explicit. This matters because causal adjustment depends on whether a variable is a confounder, mediator, or collider.

For example, if the treatment is median income:

- Region, age structure, OAC/persona mix, education, and historical area structure are plausible confounders.
- Housing prices, tenure, deprivation and insolvency may be mediators or post-treatment variables depending on the causal story.
- Including mediators estimates a **direct effect**; excluding them aims closer to a **total effect**.

## Phase 1: Bayesian hierarchical small-area estimation

The Bayesian model is the main small-area estimation model. It gives predictions with uncertainty and partial pooling.

A compact version is:

```text
log_intensity_i ~ Normal(mu_i, sigma_y)
mu_i = alpha + X_i beta + u_region[i]
u_region[r] ~ Normal(0, sigma_region)
```

For spending composition, we use the ALR-transformed shares:

```text
z_ij ~ Normal(eta_ij, sigma_j)
eta_ij = alpha_j + X_i beta_j + u_region[i,j]
```

This is a pragmatic Bayesian logistic-normal model. A fuller publication model could add spatial random effects using LAD adjacency, but region effects already give meaningful partial pooling.


````python
# Optional dependency: PyMC.
# If PyMC is unavailable, keep this cell as the model specification and use the ridge baseline below.
try:
    import pymc as pm
    import arviz as az
    PYMC_AVAILABLE = True
except Exception as exc:
    PYMC_AVAILABLE = False
    print('PyMC is not available in this environment:', exc)

# A lower-dimensional feature set stabilises Bayesian sampling with ~337 LADs.
from sklearn.decomposition import PCA

N_BAYES_COMPONENTS = min(15, X.shape[0] - 1, X.shape[1])
pca_bayes = PCA(n_components=N_BAYES_COMPONENTS, random_state=RANDOM_SEED)
X_bayes = pca_bayes.fit_transform(X)

print('Bayesian PCA matrix:', X_bayes.shape)
print('Explained variance:', round(pca_bayes.explained_variance_ratio_.sum(), 3))
````

````python
if PYMC_AVAILABLE:
    coords = {
        'lad': analysis_df['LAD16CD'].to_list(),
        'component': [f'pc_{j + 1}' for j in range(X_bayes.shape[1])],
        'region': region_codes,
        'output': Y_OUTPUT_NAMES,
    }

    try:
        import pytensor.tensor as pt
    except Exception:
        import aesara.tensor as pt

    with pm.Model(coords=coords) as bayesian_sae_model:
        X_data = pt.as_tensor_variable(X_bayes)
        region_data = pt.as_tensor_variable(region_index.astype('int64'))

        alpha = pm.Normal('alpha', 0.0, 1.0, dims='output')
        beta = pm.Normal('beta', 0.0, 0.35, dims=('component', 'output'))

        sigma_region = pm.HalfNormal('sigma_region', 0.5, dims='output')
        region_offset = pm.Normal('region_offset', 0.0, 1.0, dims=('region', 'output'))
        region_effect = pm.Deterministic(
            'region_effect', region_offset * sigma_region, dims=('region', 'output')
        )

        sigma_y = pm.HalfNormal('sigma_y', 0.5, dims='output')
        mu = alpha + pm.math.dot(X_data, beta) + region_effect[region_data]

        y_obs = pm.Normal(
            'y_obs',
            mu=mu,
            sigma=sigma_y,
            observed=Y_model,
            dims=('lad', 'output')
        )

    bayesian_sae_model
else:
    bayesian_sae_model = None
````

````python
# Run this cell when ready. Sampling can take several minutes.
# Keep target_accept high because hierarchical models often need smaller NUTS steps.
if PYMC_AVAILABLE:
    with bayesian_sae_model:
        idata_sae = pm.sample(
            draws=1000,
            tune=1000,
            chains=4,
            target_accept=0.9,
            random_seed=RANDOM_SEED,
        )
    az.summary(idata_sae, var_names=['alpha', 'sigma_region', 'sigma_y'])
else:
    idata_sae = None
````

````python
# Fast deterministic fallback/baseline for environments without PyMC.
from sklearn.linear_model import RidgeCV
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import KFold, cross_val_predict
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

ridge_sae = make_pipeline(
    StandardScaler(),
    MultiOutputRegressor(RidgeCV(alphas=np.logspace(-3, 4, 50)))
)
cv = KFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED)
Y_ridge_cv = cross_val_predict(ridge_sae, X_bayes, Y_model, cv=cv)

ridge_scores = pd.DataFrame({
    'output': Y_OUTPUT_NAMES,
    'mae': [mean_absolute_error(Y_model[:, j], Y_ridge_cv[:, j]) for j in range(Y_model.shape[1])],
    'r2': [r2_score(Y_model[:, j], Y_ridge_cv[:, j]) for j in range(Y_model.shape[1])],
})
ridge_scores
````

## Phase 2: DML causal estimates for selected treatments

Causality is not automated. A dataset does not contain causal truth by itself. Modern observational causal research usually does the following:

1. Define a treatment, outcome, unit and estimand.
2. Draw a DAG or write explicit identification assumptions.
3. Choose an adjustment set: confounders, not arbitrary predictors.
4. Estimate effects using regression, matching, inverse-probability weighting, doubly robust methods, DML, IV, or DiD.
5. Run diagnostics: overlap, balance, sensitivity, placebo outcomes, alternative adjustment sets.

For this cross-sectional LAD dataset, the safest starting design is **DML under selection-on-observables**. DML estimates nuisance functions flexibly and then estimates the effect from orthogonalised residual variation.

Plain English: compare LADs that are similar in observed confounders but differ in the treatment.


````python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.base import clone
from scipy import stats


def crossfit_predictions(estimator, X_controls, y, n_splits=5, random_state=RANDOM_SEED):
    preds = np.zeros(len(y), dtype=float)
    splitter = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    for train_idx, test_idx in splitter.split(X_controls):
        model = clone(estimator)
        model.fit(X_controls[train_idx], y[train_idx])
        preds[test_idx] = model.predict(X_controls[test_idx])
    return preds


def dml_partial_linear(df, treatment, outcome, controls, learner=None, n_splits=5):
    data = df[[treatment, outcome] + controls].dropna().copy()
    t = data[treatment].to_numpy(dtype=float)
    y = data[outcome].to_numpy(dtype=float)
    w = data[controls].to_numpy(dtype=float)

    if learner is None:
        learner = RandomForestRegressor(
            n_estimators=500,
            min_samples_leaf=8,
            random_state=RANDOM_SEED,
            n_jobs=-1,
        )

    m_hat = crossfit_predictions(learner, w, y, n_splits=n_splits)
    g_hat = crossfit_predictions(learner, w, t, n_splits=n_splits)

    y_resid = y - m_hat
    t_resid = t - g_hat

    denom = np.sum(t_resid ** 2)
    theta = np.sum(t_resid * y_resid) / denom
    residual = y_resid - theta * t_resid

    # Heteroskedasticity-robust standard error for the orthogonal score.
    score_var = np.mean((t_resid * residual) ** 2)
    se = np.sqrt(score_var / (np.mean(t_resid ** 2) ** 2 * len(data)))
    z = theta / se
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))

    return {
        'treatment': treatment,
        'outcome': outcome,
        'n': len(data),
        'theta': theta,
        'se': se,
        'ci_lower': theta - 1.96 * se,
        'ci_upper': theta + 1.96 * se,
        'p_value': p_value,
        'treatment_residual_sd': np.std(t_resid),
    }
````

````python
# Example DAG-guided adjustment set for the total effect of area income.
# This avoids post-income stress variables such as insolvency and some deprivation measures.
DML_CONTROLS_INCOME_TOTAL = [
    c for c in (
        DEMOGRAPHIC_COLUMNS + OAC_COLUMNS + COMMUTING_COLUMNS + ['Mean Age', 'Median Age']
    ) if c in analysis_df.columns
]

dml_df = pd.concat([analysis_df, region_dummies], axis=1)
DML_CONTROLS_INCOME_TOTAL += list(region_dummies.columns)
DML_CONTROLS_INCOME_TOTAL = list(dict.fromkeys([c for c in DML_CONTROLS_INCOME_TOTAL if c in dml_df.columns]))

for c in DML_CONTROLS_INCOME_TOTAL:
    dml_df[c] = pd.to_numeric(dml_df[c], errors='coerce')
dml_df[DML_CONTROLS_INCOME_TOTAL] = dml_df[DML_CONTROLS_INCOME_TOTAL].fillna(
    dml_df[DML_CONTROLS_INCOME_TOTAL].median(numeric_only=True)
)

dml_df['log_median_income'] = safe_log(dml_df[INCOME_COLUMN])

income_dml_results = []
DML_OUTCOMES = ['log_spending_intensity'] + [f'share__{c}' for c in SPENDING_COLUMNS]
for outcome in DML_OUTCOMES:
    income_dml_results.append(
        dml_partial_linear(dml_df, 'log_median_income', outcome, DML_CONTROLS_INCOME_TOTAL)
    )

income_dml_results_df = pd.DataFrame(income_dml_results).sort_values('theta', ascending=False)
income_dml_results_df
````

````python
# Direct-effect sensitivity specification:
# add housing, deprivation and labour-market controls. Differences from the total-effect
# table are substantively meaningful: they show how much of the income relationship may
# run through these area conditions.
DML_CONTROLS_INCOME_DIRECT = [
    c for c in (
        DML_CONTROLS_INCOME_TOTAL + HOUSEHOLD_COLUMNS + ECONOMIC_COLUMNS + DEPRIVATION_COLUMNS
    ) if c in dml_df.columns and c not in ['Median Income', 'Mean Income']
]
DML_CONTROLS_INCOME_DIRECT = list(dict.fromkeys(DML_CONTROLS_INCOME_DIRECT))

dml_df[DML_CONTROLS_INCOME_DIRECT] = dml_df[DML_CONTROLS_INCOME_DIRECT].apply(
    pd.to_numeric, errors='coerce'
)
dml_df[DML_CONTROLS_INCOME_DIRECT] = dml_df[DML_CONTROLS_INCOME_DIRECT].fillna(
    dml_df[DML_CONTROLS_INCOME_DIRECT].median(numeric_only=True)
)

income_dml_direct_results = []
for outcome in DML_OUTCOMES:
    income_dml_direct_results.append(
        dml_partial_linear(dml_df, 'log_median_income', outcome, DML_CONTROLS_INCOME_DIRECT)
    )

income_dml_direct_df = pd.DataFrame(income_dml_direct_results).sort_values('theta', ascending=False)
income_dml_direct_df
````

### Causal reporting discipline

Report causal results as conditional estimands, not as automatic truth.

A defensible sentence looks like:

```text
Under selection-on-observables and the stated adjustment set, a one-unit increase in log median income is associated with an estimated theta change in the outcome.
```

Then compare:

- total-effect adjustment set
- direct-effect adjustment set
- alternative learners
- leave-one-region-out estimates
- placebo outcomes or placebo treatments where possible

If results only survive one arbitrary specification, they are not strong evidence.


## Phase 3: GP multi-output benchmark

The GP is a nonlinear predictive benchmark for the same outcome vector as the Bayesian SAE model:

```text
Y = [log spending intensity, ALR composition coordinates]
```

The previous GP cell used a `SharedIndependent` kernel on unstandardised outputs. The stored results showed the decoded composition collapsing to exactly `1/12` in each category, which means the ALR latent means were effectively zero. That is a failed benchmark, not a meaningful spending model.

This version follows the GPflow multi-output notebook pattern more closely:

- standardise every output before GP training;
- use `SeparateIndependent` multi-output kernels so each output can have its own hyperparameters;
- use shared inducing inputs, as in the GPflow example;
- decode predictions only after unstandardising the latent GP mean;
- report diagnostics against the uniform-composition baseline.


````python
from sklearn.decomposition import PCA
from sklearn.metrics import mean_absolute_error, r2_score

N_GP_COMPONENTS = min(30, X.shape[0] - 1, X.shape[1])
pca_gp = PCA(n_components=N_GP_COMPONENTS, random_state=RANDOM_SEED)
X_gp = pca_gp.fit_transform(X).astype(np.float64)

Y_gp_raw = Y_model.astype(np.float64)
Y_gp_mean = Y_gp_raw.mean(axis=0, keepdims=True)
Y_gp_std = Y_gp_raw.std(axis=0, keepdims=True)
Y_gp_std[Y_gp_std == 0] = 1.0
Y_gp = ((Y_gp_raw - Y_gp_mean) / Y_gp_std).astype(np.float64)

print('GP input matrix:', X_gp.shape)
print('GP output matrix:', Y_gp.shape)
print('Explained variance:', round(pca_gp.explained_variance_ratio_.sum(), 3))
print('Standardised output mean range:', Y_gp.mean(axis=0).min(), Y_gp.mean(axis=0).max())
print('Standardised output std range:', Y_gp.std(axis=0).min(), Y_gp.std(axis=0).max())
````

````python
try:
    import gpflow as gpf
    import tensorflow as tf
    GPFLOW_AVAILABLE = True
except Exception as exc:
    GPFLOW_AVAILABLE = False
    print('GPflow is not available in this environment:', exc)

if GPFLOW_AVAILABLE:
    tf.random.set_seed(RANDOM_SEED)
    gpf.config.set_default_float(np.float64)

    N, D = X_gp.shape
    P = Y_gp.shape[1]
    M = min(80, N)
    inducing_idx = np.random.default_rng(RANDOM_SEED).choice(N, size=M, replace=False)
    Z = X_gp[inducing_idx].copy()

    # GPflow multi-output notebook pattern:
    # SeparateIndependent kernel + shared inducing locations + num_latent_gps=P.
    # This allows each transformed spending output to learn its own smoothness/variance.
    kern_list = [
        gpf.kernels.SquaredExponential(lengthscales=np.ones(D))
        + gpf.kernels.Linear(variance=0.1)
        for _ in range(P)
    ]
    kernel = gpf.kernels.SeparateIndependent(kern_list)

    inducing_variable = gpf.inducing_variables.SharedIndependentInducingVariables(
        gpf.inducing_variables.InducingPoints(Z)
    )

    gp_model = gpf.models.SVGP(
        kernel=kernel,
        likelihood=gpf.likelihoods.Gaussian(variance=0.1),
        inducing_variable=inducing_variable,
        num_latent_gps=P,
        whiten=True,
    )

    print('GPflow version:', gpf.__version__)
    print('N, D, P, M:', N, D, P, M)
    gp_model
else:
    gp_model = None
````

### Optional: correlated-output GP

`SeparateIndependent` is the most stable first benchmark and mirrors the GPflow multi-output example. It is still independent across transformed outputs conditional on the inputs. If the independent benchmark trains properly, the next academic upgrade is a **linear model of coregionalisation**:

```text
f_p(x) = sum_l W[p, l] g_l(x)
```

That lets the GP learn shared latent spending factors, such as essentials, mobility, and discretionary consumption. It is more expressive, but it should only be introduced after the simpler GP beats the uniform and ridge baselines.


### LMC GP: shared latent spending factors

The `SeparateIndependent` GP gives each transformed output its own GP, but it does not explicitly learn correlations between outputs. A **linear model of coregionalisation (LMC)** introduces a small number of shared latent GPs:

```text
f_p(x) = sum_l W[p, l] g_l(x)
```

where `p` indexes outputs and `l` indexes latent spending factors. For this problem, a small `L` is appropriate because the sample size is small. Start with `L=3` or `L=4`, corresponding loosely to latent factors such as essentials, mobility/housing, and discretionary spending.


````python
def make_lmc_gpflow_model(x_train, y_train, latent_gps=3, inducing_points=60, random_state=RANDOM_SEED):
    """Build an LMC SVGP model for standardised multi-output targets."""
    if not GPFLOW_AVAILABLE:
        raise RuntimeError('GPflow is not available.')

    N, D = x_train.shape
    P = y_train.shape[1]
    L = min(latent_gps, P)
    M = min(inducing_points, N)

    rng = np.random.default_rng(random_state)
    inducing_idx = rng.choice(N, size=M, replace=False)
    Z = x_train[inducing_idx].copy()

    # Small random W starts near independent weak mixing but lets outputs learn shared factors.
    W = rng.normal(loc=0.0, scale=0.1, size=(P, L))

    latent_kernels = [
        gpf.kernels.SquaredExponential(lengthscales=np.ones(D)) + gpf.kernels.Linear(variance=0.1)
        for _ in range(L)
    ]
    kernel = gpf.kernels.LinearCoregionalization(latent_kernels, W=W)

    inducing_variable = gpf.inducing_variables.SharedIndependentInducingVariables(
        gpf.inducing_variables.InducingPoints(Z)
    )

    model = gpf.models.SVGP(
        kernel=kernel,
        likelihood=gpf.likelihoods.Gaussian(variance=0.1),
        inducing_variable=inducing_variable,
        num_latent_gps=L,
        whiten=True,
    )
    return model


def fit_lmc_gpflow_holdout(train_idx, test_idx, n_components=30, latent_gps=3, maxiter=800, inducing_points=60):
    """Leakage-controlled holdout evaluation for LMC GPflow SVGP."""
    if not GPFLOW_AVAILABLE:
        raise RuntimeError('GPflow is not available.')

    x_train_raw = X_full.iloc[train_idx].copy()
    x_test_raw = X_full.iloc[test_idx].copy()
    y_train_raw = Y_model[train_idx].astype(np.float64)
    y_test_raw = Y_model[test_idx].astype(np.float64)

    x_mean = x_train_raw.mean(axis=0)
    x_std = x_train_raw.std(axis=0).replace(0, 1)
    x_train_scaled = ((x_train_raw - x_mean) / x_std).to_numpy(dtype=float)
    x_test_scaled = ((x_test_raw - x_mean) / x_std).to_numpy(dtype=float)

    pca = PCA(
        n_components=min(n_components, x_train_scaled.shape[0] - 1, x_train_scaled.shape[1]),
        random_state=RANDOM_SEED,
    )
    x_train = pca.fit_transform(x_train_scaled).astype(np.float64)
    x_test = pca.transform(x_test_scaled).astype(np.float64)

    y_mean = y_train_raw.mean(axis=0, keepdims=True)
    y_std = y_train_raw.std(axis=0, keepdims=True)
    y_std[y_std == 0] = 1.0
    y_train = ((y_train_raw - y_mean) / y_std).astype(np.float64)

    model = make_lmc_gpflow_model(
        x_train,
        y_train,
        latent_gps=latent_gps,
        inducing_points=inducing_points,
        random_state=RANDOM_SEED,
    )

    training_data = (x_train, y_train)
    initial_loss = model.training_loss(training_data).numpy()
    opt = gpf.optimizers.Scipy()
    opt_result = opt.minimize(
        model.training_loss_closure(training_data),
        variables=model.trainable_variables,
        method='l-bfgs-b',
        options={'maxiter': maxiter, 'disp': True},
    )
    final_loss = model.training_loss(training_data).numpy()

    pred_standardised, pred_var_standardised = model.predict_f(x_test)
    y_pred = pred_standardised.numpy() * y_std + y_mean
    metrics = spending_metrics(y_test_raw, y_pred, label=f'lmc_gpflow_L{latent_gps}')
    metrics.update({
        'latent_gps': latent_gps,
        'initial_loss': initial_loss,
        'final_loss': final_loss,
        'loss_improvement': initial_loss - final_loss,
        'optimizer_success': getattr(opt_result, 'success', None),
        'optimizer_message': str(getattr(opt_result, 'message', '')),
        'pca_explained_variance': pca.explained_variance_ratio_.sum(),
    })
    return metrics
````

## Validation protocol

The GP diagnostics above are **in-sample**. They verify that the model is no longer collapsing to the uniform composition baseline, but they are not evidence of generalisation.

For this dataset, ordinary random train/test splits are too optimistic because the expenditure labels are coarse/repeated. A stronger validation design should report at least two splits:

1. **Region-grouped validation**: tests geographic transfer.
2. **Target-profile grouped validation**: keeps identical or near-identical expenditure profiles out of both train and test, which tests whether the model learns socioeconomic structure rather than repeated labels.

All preprocessing for validation must be fit on the training fold only: scaling, PCA, output standardisation and model fitting.


````python
from sklearn.model_selection import GroupShuffleSplit, GroupKFold
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import RidgeCV
from sklearn.decomposition import PCA
from sklearn.metrics import mean_absolute_error, r2_score


def target_profile_groups(spending_frame, decimals=2):
    """Group rows with identical/near-identical raw expenditure profiles.

    The raw spending targets have relatively few unique values, so this guards against
    the same constructed target profile appearing in both train and test.
    """
    rounded = spending_frame.round(decimals).to_numpy(dtype=float)
    labels = pd.Series([tuple(row) for row in rounded]).astype('category').cat.codes.to_numpy()
    return labels


def decode_latent_matrix(y_latent):
    intensity, composition, income_shares = decode_model_outputs(y_latent)
    return intensity, composition, income_shares


def spending_metrics(y_true_latent, y_pred_latent, label='model'):
    true_intensity, true_comp, true_income_shares = decode_latent_matrix(y_true_latent)
    pred_intensity, pred_comp, pred_income_shares = decode_latent_matrix(y_pred_latent)
    uniform_comp = np.repeat(1.0 / len(SPENDING_COLUMNS), len(SPENDING_COLUMNS))

    return {
        'model': label,
        'n': len(y_true_latent),
        'log_intensity_mae': mean_absolute_error(y_true_latent[:, 0], y_pred_latent[:, 0]),
        'log_intensity_r2': r2_score(y_true_latent[:, 0], y_pred_latent[:, 0]),
        'composition_mae': np.mean(np.abs(true_comp - pred_comp)),
        'uniform_composition_mae': np.mean(np.abs(true_comp - uniform_comp)),
        'beats_uniform_composition': np.mean(np.abs(true_comp - pred_comp)) < np.mean(np.abs(true_comp - uniform_comp)),
        'income_share_mae': np.mean(np.abs(true_income_shares - pred_income_shares)),
    }


def make_holdout_split(groups, test_size=0.25, random_state=RANDOM_SEED):
    splitter = GroupShuffleSplit(n_splits=1, test_size=test_size, random_state=random_state)
    train_idx, test_idx = next(splitter.split(X_full, Y_model, groups=groups))
    return train_idx, test_idx

region_groups = analysis_df['Region'].astype('category').cat.codes.to_numpy()
profile_group_columns = SPENDING_COLUMNS + [TOTAL_SPENDING_COLUMN]
profile_groups = target_profile_groups(analysis_df[profile_group_columns])

print('Rows:', len(analysis_df))
print('Regions:', len(np.unique(region_groups)))
print('Unique target profiles:', len(np.unique(profile_groups)))
print('Largest target-profile group sizes:')
display(pd.Series(profile_groups).value_counts().head(10))
````

````python
def evaluate_ridge_holdout(train_idx, test_idx, n_components=30):
    """Fast leakage-controlled benchmark: train-only scaling/PCA/output fit."""
    x_train_raw = X_full.iloc[train_idx].copy()
    x_test_raw = X_full.iloc[test_idx].copy()
    y_train = Y_model[train_idx]
    y_test = Y_model[test_idx]

    x_mean = x_train_raw.mean(axis=0)
    x_std = x_train_raw.std(axis=0).replace(0, 1)
    x_train_scaled = ((x_train_raw - x_mean) / x_std).to_numpy(dtype=float)
    x_test_scaled = ((x_test_raw - x_mean) / x_std).to_numpy(dtype=float)

    pca = PCA(n_components=min(n_components, x_train_scaled.shape[0] - 1, x_train_scaled.shape[1]), random_state=RANDOM_SEED)
    x_train = pca.fit_transform(x_train_scaled)
    x_test = pca.transform(x_test_scaled)

    model = make_pipeline(
        StandardScaler(),
        MultiOutputRegressor(RidgeCV(alphas=np.logspace(-3, 4, 50)))
    )
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    return spending_metrics(y_test, y_pred, label='ridge')

region_train_idx, region_test_idx = make_holdout_split(region_groups, test_size=0.25)
profile_train_idx, profile_test_idx = make_holdout_split(profile_groups, test_size=0.25)

ridge_validation = pd.DataFrame([
    evaluate_ridge_holdout(region_train_idx, region_test_idx) | {'split': 'region_group_holdout'},
    evaluate_ridge_holdout(profile_train_idx, profile_test_idx) | {'split': 'target_profile_group_holdout'},
])
ridge_validation
````

````python
def fit_gpflow_holdout(train_idx, test_idx, n_components=30, maxiter=800, inducing_points=60):
    """Leakage-controlled GP holdout evaluation.

    This is intentionally a single holdout utility because fitting GPflow inside many CV
    folds is expensive. Use it after the fast ridge validation to see whether the GP is
    earning its complexity.
    """
    if not GPFLOW_AVAILABLE:
        raise RuntimeError('GPflow is not available.')

    x_train_raw = X_full.iloc[train_idx].copy()
    x_test_raw = X_full.iloc[test_idx].copy()
    y_train_raw = Y_model[train_idx].astype(np.float64)
    y_test_raw = Y_model[test_idx].astype(np.float64)

    x_mean = x_train_raw.mean(axis=0)
    x_std = x_train_raw.std(axis=0).replace(0, 1)
    x_train_scaled = ((x_train_raw - x_mean) / x_std).to_numpy(dtype=float)
    x_test_scaled = ((x_test_raw - x_mean) / x_std).to_numpy(dtype=float)

    pca = PCA(n_components=min(n_components, x_train_scaled.shape[0] - 1, x_train_scaled.shape[1]), random_state=RANDOM_SEED)
    x_train = pca.fit_transform(x_train_scaled).astype(np.float64)
    x_test = pca.transform(x_test_scaled).astype(np.float64)

    y_mean = y_train_raw.mean(axis=0, keepdims=True)
    y_std = y_train_raw.std(axis=0, keepdims=True)
    y_std[y_std == 0] = 1.0
    y_train = ((y_train_raw - y_mean) / y_std).astype(np.float64)

    N, D = x_train.shape
    P = y_train.shape[1]
    M = min(inducing_points, N)
    inducing_idx = np.random.default_rng(RANDOM_SEED).choice(N, size=M, replace=False)
    Z = x_train[inducing_idx].copy()

    kern_list = [
        gpf.kernels.SquaredExponential(lengthscales=np.ones(D)) + gpf.kernels.Linear(variance=0.1)
        for _ in range(P)
    ]
    kernel = gpf.kernels.SeparateIndependent(kern_list)
    inducing_variable = gpf.inducing_variables.SharedIndependentInducingVariables(
        gpf.inducing_variables.InducingPoints(Z)
    )
    model = gpf.models.SVGP(
        kernel=kernel,
        likelihood=gpf.likelihoods.Gaussian(variance=0.1),
        inducing_variable=inducing_variable,
        num_latent_gps=P,
        whiten=True,
    )

    training_data = (x_train, y_train)
    initial_loss = model.training_loss(training_data).numpy()
    opt = gpf.optimizers.Scipy()
    opt_result = opt.minimize(
        model.training_loss_closure(training_data),
        variables=model.trainable_variables,
        method='l-bfgs-b',
        options={'maxiter': maxiter, 'disp': True},
    )
    final_loss = model.training_loss(training_data).numpy()

    pred_standardised, pred_var_standardised = model.predict_f(x_test)
    y_pred = pred_standardised.numpy() * y_std + y_mean
    metrics = spending_metrics(y_test_raw, y_pred, label='gpflow_svgp')
    metrics.update({
        'initial_loss': initial_loss,
        'final_loss': final_loss,
        'loss_improvement': initial_loss - final_loss,
        'optimizer_success': getattr(opt_result, 'success', None),
        'optimizer_message': str(getattr(opt_result, 'message', '')),
        'pca_explained_variance': pca.explained_variance_ratio_.sum(),
    })
    return metrics

# Run one or both. They can take a while.
gp_region_validation = fit_gpflow_holdout(region_train_idx, region_test_idx, maxiter=800)
gp_profile_validation = fit_gpflow_holdout(profile_train_idx, profile_test_idx, maxiter=800)
pd.DataFrame([gp_region_validation | {'split': 'region_group_holdout'},
              gp_profile_validation | {'split': 'target_profile_group_holdout'}])
````

### How to read validation results

A useful GP should satisfy all of these:

- predicted composition rows sum to one after decoding;
- out-of-sample composition MAE beats the uniform-composition baseline;
- out-of-sample metrics beat or match the ridge benchmark enough to justify GP complexity;
- results remain reasonable under region-grouped and target-profile-grouped splits;
- optimiser loss falls substantially, even if L-BFGS occasionally hits the iteration limit.

If the GP only wins in-sample but loses on target-profile holdout, it is learning the repeated target construction rather than a generalisable socioeconomic spending surface.


## Model comparison and academic reporting

The paper-style comparison should include:

1. **Data construction**: LAD unit, expenditure source, OAC source, missingness, duplicated/coarse spending labels.
2. **Outcome validity**: composition sums to one; spending intensity separated from composition.
3. **Bayesian SAE**: posterior intervals, region-level partial pooling, posterior predictive checks.
4. **DML causality**: estimand, DAG/adjustment set, overlap diagnostics, sensitivity to direct vs total controls.
5. **GP benchmark**: predictive performance against ridge/elastic net, Bayesian posterior mean, and the uniform-composition baseline.
6. **Validation**: grouped or region-aware CV. Avoid naive random splits if repeated/coarse target labels dominate.
7. **Limitations**: ecological inference, observational confounding, spending labels are small-area estimates rather than directly observed household records.

A strong abstract claim would be:

We estimate LAD-conditioned household spending intensity and expenditure composition using a hierarchical small-area model, assess selected income/deprivation effects with double machine learning, and benchmark nonlinear multi-output prediction with Gaussian processes.

### Page 2: Bayesian Hierarchical Small-Area Estimation

**Message for slide:** Small-Area Estimation (SAE) is used when reliable estimates are needed for local geographies where observations are limited or noisy. Here, each Local Authority District (LAD) is treated as a small area. The aim is to estimate an area-level household spending profile while borrowing strength across socioeconomically similar LADs and across regions.

The Bayesian hierarchical model provides three advantages:

- It produces uncertainty-aware estimates, not only point predictions.
- It partially pools LAD estimates through regional random effects.
- It stabilises estimates for areas with weaker or noisier expenditure signal.

Let $i = 1,\dots,N$ index LADs and $j = 1,\dots,J$ index modelled spending outcomes. The outcome vector is:

$$
\mathbf{y}_i =
\left[
\log\left(\frac{\text{total expenditure}_i}{\text{median income}_i}\right),
\operatorname{ALR}(\mathbf{s}_i)
\right]
$$

where $\mathbf{s}_i$ is the vector of expenditure category shares and $\operatorname{ALR}(\cdot)$ is the additive log-ratio transform used to model compositional spending shares.

For each outcome $j$:

$$
y_{ij} \sim \mathcal{N}(\mu_{ij}, \sigma_j^2)
$$

$$
\mu_{ij}
=
\alpha_j
+
\mathbf{x}_i^\top \boldsymbol{\beta}_j
+
u_{r[i]j}
$$

$$
u_{rj} \sim \mathcal{N}(0, \tau_j^2)
$$

where $\mathbf{x}_i$ contains socioeconomic predictors, $r[i]$ denotes the region of LAD $i$, $u_{r[i]j}$ is a region-level random effect, $\sigma_j^2$ captures residual outcome variation, and $\tau_j^2$ captures between-region variation.

**Interpretation:**  
The model estimates LAD spending profiles by combining local socioeconomic evidence with regional partial pooling. This reduces overfitting to noisy local patterns while preserving systematic geographic differences in expenditure behaviour.

### Page 3: Double Machine Learning (DML) Causality

**Message for slide:** The Bayesian SAE model estimates area-level spending profiles and uncertainty. DML is a complementary step: it asks whether a specific socioeconomic variable, such as median income, has an interpretable conditional association with spending outcomes after adjusting for observed confounders.

The causal question is not automated by the data. We first define a treatment, outcome and adjustment set. Here, the treatment is area income and the outcomes are spending intensity and expenditure category shares.

Let $T_i$ denote the treatment for LAD $i$:

$$
T_i = \log(\text{median income}_i)
$$

Let $Y_i$ denote a spending outcome, such as log spending intensity or a category budget share. Let $\mathbf{W}_i$ denote observed adjustment variables, including OAC composition, demographic structure, commuting patterns and region.

DML uses two predictive nuisance models:

$$
m(\mathbf{W}_i) = \mathbb{E}[Y_i \mid \mathbf{W}_i]
$$

$$
g(\mathbf{W}_i) = \mathbb{E}[T_i \mid \mathbf{W}_i]
$$

The outcome and treatment are residualised:

$$
\tilde{Y}_i = Y_i - \hat{m}(\mathbf{W}_i)
$$

$$
\tilde{T}_i = T_i - \hat{g}(\mathbf{W}_i)
$$

The treatment effect is then estimated from the orthogonalised relationship:

$$
\tilde{Y}_i = \theta \tilde{T}_i + \varepsilon_i
$$

where $\theta$ is the estimated conditional effect of income on the spending outcome.

**Connection to Bayesian SAE:**  
The Bayesian SAE model estimates the local spending surface and its uncertainty. DML then interrogates this spending structure by estimating how much of the variation in each spending outcome is associated with income, after adjusting for observed area composition. In this sense, SAE answers “what is the expected spending profile of each area?”, while DML asks “how does income relate to that profile, conditional on observed confounders?”

**Interpretation of results:**  
The DML estimates suggest that higher median income is associated with higher shares of recreation, restaurants and hotels, transport, education, and clothing. It is associated with lower shares of housing, food, communication and miscellaneous goods. This pattern is consistent with an Engel-curve interpretation: as income rises, essential categories take a smaller share of spending while discretionary categories take a larger share.

These estimates should be interpreted as conditional causal evidence under the stated adjustment assumptions, not as automatic causal discovery.

### Page 4: Multi-Output Sparse Variational Gaussian Process

**Message for slide:** The Bayesian SAE model provides uncertainty-aware small-area estimates, while DML estimates interpretable income effects under causal assumptions. The Sparse Variational Gaussian Process (SVGP) model is the predictive layer: it learns a flexible nonlinear mapping from LAD socioeconomic context to household spending behaviour.

The SVGP predicts the same outcome representation used throughout the analysis:

$$
\mathbf{y}_i =
\left[
\log\left(\frac{\text{total expenditure}_i}{\text{median income}_i}\right),
\operatorname{ALR}(\mathbf{s}_i)
\right]
$$

where $\mathbf{s}_i$ is the vector of expenditure category shares. The inverse ALR transform maps predictions back to valid spending shares that are positive and sum to one.

Let $\mathbf{x}_i$ denote the socioeconomic feature vector for LAD $i$. The model assumes:

$$
\mathbf{y}_i = \mathbf{f}(\mathbf{x}_i) + \boldsymbol{\varepsilon}_i
$$

where $\mathbf{f}(\cdot)$ is a vector-valued latent function and $\boldsymbol{\varepsilon}_i$ is observation noise.

Because full GP inference can be computationally expensive, we use a sparse variational approximation with inducing points:

$$
\mathbf{u}_j = f_j(\mathbf{Z})
$$

where $\mathbf{Z}$ is a set of inducing locations. The variational distribution approximates the posterior over latent function values:

$$
q(\mathbf{u}_j) \approx p(\mathbf{u}_j \mid \mathbf{y})
$$

The main benchmark uses separate latent functions for each transformed output:

$$
f_j(\mathbf{x}) \sim \mathcal{GP}(0, k_j(\mathbf{x}, \mathbf{x'}))
$$

for outcomes $j = 1,\dots,J$. This allows each spending outcome to have its own nonlinear relationship with socioeconomic context.

We also test a Linear Model of Coregionalisation (LMC), which allows outputs to share latent spending factors:

$$
f_j(\mathbf{x}) =
\sum_{\ell=1}^{L}
w_{j\ell} g_{\ell}(\mathbf{x})
$$

$$
g_{\ell}(\mathbf{x}) \sim \mathcal{GP}(0, k_{\ell}(\mathbf{x}, \mathbf{x'}))
$$

where $g_{\ell}(\cdot)$ are shared latent GPs and $w_{j\ell}$ are output-specific loadings. In this study, LMC tests whether categories such as essentials, housing/mobility and discretionary spending share common latent structure.

**Connection to previous steps:**  
The Bayesian SAE model gives a principled small-area estimation framework and quantifies uncertainty. DML provides interpretable income-effect estimates under explicit adjustment assumptions. The SVGP model then asks a predictive question: given the full socioeconomic profile of a LAD, how accurately can we predict its spending intensity and expenditure composition?

**Validation:**  
The SVGP is evaluated using grouped holdout splits rather than only in-sample fit. Region-grouped validation tests geographic transfer, while target-profile grouped validation checks whether the model generalises beyond repeated expenditure profiles.

**Interpretation of results:**  
The separate-output SVGP provides the most stable spending-composition predictions. The LMC-SVGP improves regional transfer for spending intensity, suggesting that shared latent factors help capture broad geographic structure, but it weakens composition prediction in the current run. Therefore, LMC is useful as an academic extension, while the separate-output SVGP remains the main predictive benchmark.

````python
# Slide 4b: SHAP interpretation for the fitted separate-output GP.
# This explains GP predictions, not causal effects.

try:
    import shap
except ImportError as exc:
    raise ImportError(
        "Install SHAP first, e.g. `pip install shap` or "
        "`conda install -c conda-forge shap`."
    ) from exc

feature_names = X_full.columns.to_list()

# Keep this small: model-agnostic SHAP is expensive.
background = shap.sample(X_full, 40, random_state=RANDOM_SEED)
explain_X = shap.sample(X_full, 80, random_state=RANDOM_SEED)

discretionary_categories = ['Recreation & culture', 'Restaurants & hotels', 'Transport']
discretionary_idx = [SPENDING_COLUMNS.index(c) for c in discretionary_categories]

def _as_feature_frame(x):
    frame = pd.DataFrame(x, columns=feature_names)
    frame = frame.apply(pd.to_numeric, errors='coerce')
    frame = frame.fillna(X_mean)
    frame = frame.infer_objects(copy=False)
    return frame

def gp_predict_log_intensity(x):
    frame = _as_feature_frame(x)
    pred = predict_gp_from_feature_frame(frame)
    return np.log(pred['spending_intensity'])

def gp_predict_discretionary_share(x):
    frame = _as_feature_frame(x)
    pred = predict_gp_from_feature_frame(frame)
    return pred['composition_shares'][:, discretionary_idx].sum(axis=1)

def shap_bar_plot(shap_values, feature_names, title, filename, max_display=15):
    mean_abs = np.abs(shap_values).mean(axis=0)
    order = np.argsort(mean_abs)[-max_display:]

    fig, ax = plt.subplots(figsize=(8.5, 6.0))
    ax.barh(np.array(feature_names)[order], mean_abs[order], color='#31688e')
    ax.set_xlabel('Mean absolute SHAP value')
    ax.set_title(title)
    fig.tight_layout()
    save_slide_fig(fig, filename)
    plt.show()

# 1. Explain predicted log spending intensity.
explainer_intensity = shap.KernelExplainer(
    gp_predict_log_intensity,
    background.to_numpy()
)
shap_intensity = explainer_intensity.shap_values(
    explain_X.to_numpy(),
    nsamples=250
)

shap_bar_plot(
    shap_intensity,
    feature_names,
    'SHAP: drivers of GP-predicted spending intensity',
    'slide_4b_shap_gp_log_intensity.png'
)

# 2. Explain predicted discretionary composition share.
explainer_discretionary = shap.KernelExplainer(
    gp_predict_discretionary_share,
    background.to_numpy()
)
shap_discretionary = explainer_discretionary.shap_values(
    explain_X.to_numpy(),
    nsamples=250
)

shap_bar_plot(
    shap_discretionary,
    feature_names,
    'SHAP: drivers of GP-predicted discretionary share',
    'slide_4b_shap_gp_discretionary_share.png'
)

# Export SHAP summary tables for PowerPoint.
shap_intensity_table = pd.DataFrame({
    'feature': feature_names,
    'mean_abs_shap': np.abs(shap_intensity).mean(axis=0),
}).sort_values('mean_abs_shap', ascending=False)

shap_discretionary_table = pd.DataFrame({
    'feature': feature_names,
    'mean_abs_shap': np.abs(shap_discretionary).mean(axis=0),
}).sort_values('mean_abs_shap', ascending=False)

display_table(shap_intensity_table.head(20), 'slide_4b_shap_gp_log_intensity.csv')
display_table(shap_discretionary_table.head(20), 'slide_4b_shap_gp_discretionary_share.csv')
````
